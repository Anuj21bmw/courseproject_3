```python
#renaming columns in matches
df2.rename(columns={"match_number":"match_id", "start_date":"date"}, inplace=True)
```

```python
df1.isna().sum()
```

```python
#creating a wicket column to get boolean of wicket fall
df1["wicket"] = df1["player_dismissed"].apply(lambda x: 1 if isinstance(x, str) else 0)
df1
```

```python
#filling na values
df1[["wides", "noballs", "byes", "legbyes", "penalty"]] = df1[["wides", "noballs", "byes", "legbyes", "penalty"]].fillna(0)
```

```python
df1.wicket_type.fillna("no wicket", inplace=True)
df1.player_dismissed.fillna("no wicket", inplace=True)
```

```python
df1.isna().sum()
```

```python
df2.isna().sum()
```

```python
df2.rename(columns={"winner_runs":"win_by_runs", "winner_wickets":"win_by_wickets"}, inplace=True)
df2.win_by_runs.fillna(0, inplace=True)
df2.win_by_wickets.fillna(0, inplace=True)
df2.reserve_umpire.fillna("unknown", inplace=True)
```

```python
df2.isna().sum()
```

```python
#creating over and balls from ball column which is in 0.1, 0.2 format for first over
df1["ball"] = df1["ball"].astype(str)
df1[["over", "ball_num"]] = df1["ball"].str.split(".", expand=True).astype(int)
df1["ball"] = df1["ball"].astype(float)
df1["over"] = df1["over"] + 1
df1["ball_left"] = 306 - (df1["over"]*6 + df1["ball_num"])
```

```python
#creating total runs column to calculate runs on each ball incusive of extras
df1["total_runs"] = df1["runs_off_bat"] + df1["extras"]
```

```python
#creating innings df so we can merger total runs for each innings with in original df
innings = df1.groupby(["match_id", "innings"])["total_runs"].sum().reset_index()
innings.rename(columns={"total_runs":"innings_total"}, inplace=True)
```

```python
deliveries = df1.merge(innings, on=["match_id", "innings"])
deliveries
```

```python
#calculating cumulative runs
deliveries["cumulative_runs"] = deliveries.groupby(["match_id", "innings"])["total_runs"].cumsum()
deliveries
```

```python
#creating run rate column
deliveries["run_rate"] = deliveries.groupby(["match_id","innings"]).apply(lambda x: (x["cumulative_runs"]*6)/(300-x["ball_left"])).reset_index(level=[0,1], drop=True)
deliveries
```

```python
#filtering for innings1 and 2 so we can add some columns respectively
innings_1 = deliveries[deliveries["innings"] == 1]

innings_1 = innings_1.groupby("match_id")["cumulative_runs"].max().reset_index()
innings_1.rename(columns={"cumulative_runs": "target"}, inplace=True)

innings_2 = deliveries[deliveries["innings"] == 2]
innings_2 = innings_2.merge(innings_1, on="match_id", how="left")
innings_2["target"] = innings_2["target"]+1
```

```python
#adding runs needed and rrr to innings 2
innings_2["runs_needed"] = innings_2["target"] - innings_2["cumulative_runs"]
innings_2["required_run_rate"] = (innings_2["runs_needed"]*6) / innings_2["ball_left"]
```

```python
#adding wickets remaining to innings 2
innings_2["wickets_remaining"] = innings_2.groupby("match_id")["wicket"].apply(lambda x: 10 - x.cumsum()).reset_index(level=0, drop=True)
#innings 1 data
innings_1_f = deliveries[deliveries["innings"] == 1]
#adding wickets remaining
innings_1_f["wickets_remaining"] = innings_1_f.groupby("match_id")["wicket"].apply(lambda x: 10 - x.cumsum()).reset_index(level=0, drop=True)
```

```python
#concatinating innings1,2
innings = pd.concat([innings_1_f,innings_2])
innings.isna().sum()
```

```python
len(innings)
```

```python
innings.fillna(0, inplace=True)
innings = innings[~((innings.required_run_rate > 500) | (innings.run_rate > 100))]


#merging with matches data to get winners and other data that can help our model
innings = innings.merge(df2[["match_id","winner", "win_by_runs", "win_by_wickets"]], on="match_id")
```

```python
df2.columns
```

```python
#adding result column
innings["result"] = innings.apply(lambda x: 1 if x["batting_team"] == x["winner"] else 0, axis=1)
#making the data random to avoid overfitting
innings = innings.sample(frac=1)
```

```python
#defining model data for prediction
model_column = innings[["venue", "batting_team", "bowling_team", "ball", "innings_total", "wickets_remaining", "run_rate", "required_run_rate",
                      "ball_left", "runs_needed", "win_by_runs", "win_by_wickets", "target", "result"]]
```

```python
model_column.isna().sum()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
model_column.dtypes
```

```python
#differentiating data from result column
X = model_column.iloc[:,:-1]
y = model_column.iloc[:, -1]
```

```python
#splitting data to fit into model
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100)
```

```python
#defining columns to encode and scale
ohe_cols = [col for col in X.columns if X[col].dtypes not in (int, float)]
scale_cols = [col for col in X.columns if X[col].dtypes in (int, float)]
ohe_cols, scale_cols
```

```python
#column transformer
ct = make_column_transformer(
    (OneHotEncoder(sparse=False, drop="first"), ohe_cols),
    remainder="passthrough"
)
```

```python
#defining pipeline and steps
pipe = make_pipeline(
    ct,
    LogisticRegression(solver="liblinear")
)
```

```python
#fitting into model
pipe.fit(x_train, y_train)
```

```python
#testing the model and its accuracy
pred = pipe.predict(x_test)
acc = accuracy_score(y_test, pred)
acc
```

```python
pipe.predict_proba(x_test)[9]
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

# After fitting the model, calculate additional metrics
precision = precision_score(y_test, pred)
recall = recall_score(y_test, pred)
f1 = f1_score(y_test, pred)

print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)



#Precision: 0.959711136450019
#Recall: 0.9549924357034796
#F1 Score: 0.9573459715639812
```
